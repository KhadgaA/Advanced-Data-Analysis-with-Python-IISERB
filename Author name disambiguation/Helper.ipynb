```python

personID = {}
paperid = []
# title = []
# organization = []

df = pd.DataFrame()

data_2 = {}

stopword = ['at','based','in','of','for','on','and','to','an','using','with','the']
for fname in file_to_open:
    f = open(data_folder / fname,'r',encoding = 'utf-8').read()
#     print(f)
    text=re.sub(u"&",u"and",f)
    
#     print(text)
    root = ET.fromstring(text)
    data = {}
    title = []
    year = []
    authors = []
    jconf = []
    pid = []
    label = []
    
    
    for i in root.findall('publication'):
        t1 = i.find('title').text
        for word in stopword: 
            if word in t1:
                t1 = re.sub("\s"+word+"\s"," ",t1)
                
        title.append(t1)
        year.append(i.find('year').text)
        authors.append(i.find("authors").text)
        jconf.append(i.find("jconf").text)
        pid.append(i.find('id').text)
        label.append(i.find('label').text)
   
    Name = root.find('FullName').text
    personid = root.find("personID").text
    personID[personid] = Name

    data = {
        
        'author' :Name,
        'title':title,
        'year':year,
        'authors':authors,
        'jconf':jconf,
        'pid':pid,
        'label':label
        
    }
    
        
    data_all = pd.DataFrame(data)
#     dataf  = pd.DataFrame.from_dict(data)
#     data_2[Name] = dataf
    
    
    
#     print(data_2)
    df = df.append(data_all)   
# df.set_index('year')
# df = pd.DataFrame.from_dict(data_2)
print(df)
# print(data_2)
        
```
